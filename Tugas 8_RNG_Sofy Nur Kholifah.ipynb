```python

import random

def simulasi_skor_tes(jumlah_siswa, skor_min, skor_max):
    """
    Simulasi skor tes siswa.

    Args:
        jumlah_siswa (int): Jumlah siswa yang mengikuti tes.
        skor_min (int): Skor minimum yang mungkin.
        skor_max (int): Skor maksimum yang mungkin.
    """
    print("=== Simulasi Skor Tes ===")
    skor_siswa = []

    for i in range(1, jumlah_siswa + 1):
        # Generate skor acak untuk siswa
        skor = random.randint(skor_min, skor_max)
        skor_siswa.append(skor)
        print(f"Siswa {i}: Skor = {skor}")

    # Statistik hasil tes
    rata_rata = sum(skor_siswa) / jumlah_siswa
    skor_tertinggi = max(skor_siswa)
    skor_terendah = min(skor_siswa)

    print("\n=== Statistik Hasil Tes ===")
    print(f"Jumlah Siswa: {jumlah_siswa}")
    print(f"Rata-rata Skor: {rata_rata:.2f}")
    print(f"Skor Tertinggi: {skor_tertinggi}")
    print(f"Skor Terendah: {skor_terendah}")

# Parameter simulasi
jumlah_siswa = 15  # Jumlah siswa diubah menjadi 15
skor_min = 0       # Skor minimum
skor_max = 100     # Skor maksimum

simulasi_skor_tes(jumlah_siswa, skor_min, skor_max)
```
